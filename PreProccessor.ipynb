```python
import numpy as np
import pandas as pd


df3=pd.read_csv("data2/3.csv",header=None)
df4=pd.read_csv("data2/4.csv",header=None)
df5=pd.read_csv("data2/5.csv",header=None)
df6=pd.read_csv("data2/6.csv",header=None)

frames=[df3,df4,df5,df6]
df = pd.concat(frames)
# df.to_csv(r'data2/merged.csv', index = False)
# print(df6)
```

```python


colnames=['ip_address','subnet_mask','packet_count','byte_count','received_packets_1','transmitted_packets_1','received_bytes_1','transmitted_bytes_1','received_packets_2','transmitted_packets_2','received_bytes_2','transmitted_bytes_2','received_packets_3','transmitted_packets_3','received_bytes_3','transmitted_bytes_3']
df=pd.read_csv("data2/merged.csv",names=colnames,dtype={'ip_address':str,'subnet_mask':str})
df = df.iloc[1:]
df['ip_address'] = df['ip_address'].astype('str') 

df=df.sort_values('ip_address')
df = df.reset_index(drop=True)


# df.to_csv(r'data2/merged_sorted.csv', index = False)

df['packet_count'] = pd.to_numeric(df['packet_count'],errors='coerce')
df['byte_count'] = pd.to_numeric(df['byte_count'],errors='coerce')

df['received_packets_1'] = pd.to_numeric(df['received_packets_1'],errors='coerce')
df['transmitted_packets_1'] = pd.to_numeric(df['transmitted_packets_1'],errors='coerce')
df['received_bytes_1'] = pd.to_numeric(df['received_bytes_1'],errors='coerce')
df['transmitted_bytes_1'] = pd.to_numeric(df['transmitted_bytes_1'],errors='coerce')

df['received_packets_2'] = pd.to_numeric(df['received_packets_2'],errors='coerce')
df['transmitted_packets_2'] = pd.to_numeric(df['transmitted_packets_2'],errors='coerce')
df['received_bytes_2'] = pd.to_numeric(df['received_bytes_2'],errors='coerce')
df['transmitted_bytes_2'] = pd.to_numeric(df['transmitted_bytes_2'],errors='coerce')

df['received_packets_3'] = pd.to_numeric(df['received_packets_3'],errors='coerce')
df['transmitted_packets_3'] = pd.to_numeric(df['transmitted_packets_3'],errors='coerce')
df['received_bytes_3'] = pd.to_numeric(df['received_bytes_3'],errors='coerce')
df['transmitted_bytes_3'] = pd.to_numeric(df['transmitted_bytes_3'],errors='coerce')

res=df.copy()

res.packet_count=res.packet_count.diff()
res.byte_count=res.byte_count.diff()

res.received_packets_1=res.received_packets_1.diff()
res.received_packets_2=res.received_packets_2.diff()
res.received_packets_3=res.received_packets_3.diff()

res.transmitted_packets_1=res.transmitted_packets_1.diff()
res.transmitted_packets_2=res.transmitted_packets_2.diff()
res.transmitted_packets_3=res.transmitted_packets_3.diff()

res.received_bytes_1=res.received_bytes_1.diff()
res.received_bytes_2=res.received_bytes_2.diff()
res.received_bytes_3=res.received_bytes_3.diff()

res.transmitted_bytes_1=res.transmitted_bytes_1.diff()
res.transmitted_bytes_2=res.transmitted_bytes_2.diff()
res.transmitted_bytes_3=res.transmitted_bytes_3.diff()

print(res)

```
